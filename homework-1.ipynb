任务：

提取SARS-Cov-2样本占比大于万分之一的个Lineage 75%和10%水平的特征突变，以及mutation的总数

ps：Lineage 75%和10%水平的特征突变：该Lineage 75%和10%以上的毒株都具有的突变

以下为代码流程：

第一步：将本次实验的相关数据读取到python的工作环境当中

```python
import pandas as pd
df = pd.read_csv('C:/Users/20156/Desktop/variant_surveillance.zip', delimiter='\t').dropna(axis=0, subset=['AA Substitutions', 'Pango lineage'])
df.head()
```

第二步：计算每一个Lineage所占的比例

```python
counts_pl = df['Pango lineage'].value_counts(normalize=True)
print(counts_pl)
```

第三步：将符合条件（比例超过万分之一的Lineage）放入新的变量中

```python
mask_pl = df['Pango lineage'].isin(counts_pl[counts_pl >= 478].index)
df_filtered = df.groupby('Pango lineage').filter(lambda x: x['Pango lineage'].count() >= 478)
```

```python
df_filtered['Pango lineage'].value_counts(normalize=True)
```

```python
df_filtered['Pango lineage'].value_counts()
```

于是我们得到了比例超过万分之一的Lineage，并且用了新的变量。

第四步：考虑到AA Substitutions每一组都是字符串类型，所以将其转化为列表。为了更好处理，用for循环，将所有的AA Substitutions的内容都记录到一个列表里面

```python
df_filtered['AA Substitutions'].value_counts()
```

```python
totel_aa=[aa for record in [*df_filtered['AA Substitutions']] for aa in record.strip('(').strip(')').split(',') ]
totel_aa
```

为了更方便处理，我们将列表转化为DataFrame的格式

```python
df_totelaa = pd.DataFrame(totel_aa)
```

第五步：此时我们有了包含所有AA Substitutions的内容的DataFrame格式的变量。因为每个病毒关于一个变异的记录是唯一的，所以在列表中的变异的个数即为含有这种Lineage的数目，所以求比例，只需要用当前数目÷总数目即可。并且记录在result中

```python
df_totelaa.value_counts()
type(df_totelaa)
result = df_totelaa.value_counts().apply(lambda x: x/4767972)
```

此时注意到result的数据类型已经改变，为了方便后续处理，需要将其转化为DataFrame格式

```python
type(result)
result2 = result.to_frame().rename(columns={0:'bizhi'})
result2
```

然后再次基础上，提取出75%水平

```python
result2.loc[result2['bizhi'] >0.75]
```

提取出10%水平

```python
result2.loc[result2['bizhi'] >0.1]
```

最后的数量输出结果包含，试验结束。
